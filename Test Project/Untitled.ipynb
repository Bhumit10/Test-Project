```python
train_df = pd.read_csv(r'C:\Users\LAPSHOP.IN\Downloads\Python\Hackathon\train.csv', header=0, index_col=0)
test_df = pd.read_csv(r'C:\Users\LAPSHOP.IN\Downloads\Python\Hackathon\test.csv', header=0, index_col=0)
train_df.head()
```

```python
train_df.describe()
```

```python
train_df.drop(["Saving Account"], axis = 1, inplace = True)
test_df.drop(["Saving Account"], axis = 1, inplace = True)
```

```python
train_df["account_info"].fillna(train_df["account_info"].mode()[0], inplace=True)
print(train_df.isnull().sum())
test_df["account_info"].fillna(test_df["account_info"].mode()[0], inplace=True)
print(test_df.isnull().sum())
```

```python
train_df.columns
```

```python
train_df = train_df[['account_info', 'duration_month', 'Loan', 'Regularity', 'Purpose',
                     'savings_account', 'employment_st', 'Salary',
       'investment', 'Gender', 'Status', 'gurantors', 'resident_since',
       'property_type', 'age', 'installment_type', 'housing_type',
       'credits_no', 'job_type', 'liables', 'telephone', 'foreigner','credit_amount']]
```

```python
train_df.info()
```

```python
test_df.info()
```

```python
colname =[]
for x in train_df.columns:
    if train_df[x].dtype == 'object':
        colname.append(x)

colname
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for x in colname:
    train_df[x] = le.fit_transform(train_df[x])
```

```python
train_df.head()
```

```python
colname1 = []
for x in test_df.columns:
    if test_df[x].dtype == 'object':
        colname1.append(x)
        
colname1
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for x in colname1:
    test_df[x] = le.fit_transform(test_df[x])
```

```python
test_df.head()
```

```python
X = train_df.values[:,:-1]
Y = train_df.values[:,-1]
Y = Y.astype(int)
```

```python
Y.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(X)
X = scaler.transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=10)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor

randomregressor = RandomForestRegressor(n_estimators=100, random_state=10)

randomregressor.fit(X_train, Y_train)
Y_pred = randomregressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error

print(r2_score(Y_test, Y_pred))
print(mean_squared_error(Y_test, Y_pred))
```

```python
print(list(zip(train_df.columns[:-1], randomregressor.feature_importances_)))
```

```python
test = test_df.values
test = scaler.transform(test)
```

```python
from sklearn.ensemble import RandomForestRegressor

randomregressor = RandomForestRegressor(n_estimators=100, random_state=0)

randomregressor.fit(X_train, Y_train)

Y_pred_new = randomregressor.predict(test)
```
